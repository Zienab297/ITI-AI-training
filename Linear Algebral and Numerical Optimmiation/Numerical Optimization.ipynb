```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('RegData.csv',delimiter=',')
```

```python
data
```

```python
X = data[:,0]
```

```python
X
```

```python
y = data[:,1]
```

```python
y
```

```python
plt.figure(figsize =(7,7))
plt.xlabel("X")
plt.ylabel("y")
plt.scatter(X,y)
```

```python
#intialize theta_0 ,theta_1
theta_0 = 3
theta_1 = 0
alpha =0.001
m = X.shape[0]

print("the no of sample of data (m):",m)
```

```python
h = theta_0 +theta_1 * X
print(f'h(x)= {h}')
```

```python
e = h - y #h is y predicted
j = np.linalg.norm(e)**2/(2*m)
print(f'error vector \n {e}\n')
print(f'cost function=: {j}')
```

```python
#calculate gradient
d_theta_0 = e.sum()/m
d_theta_1 = (e@X)/m

print(f'd_theta_0:{d_theta_0}')
print(f'd_theta_1:{d_theta_1}')
```

```python
theta_0 = theta_0 -alpha * d_theta_0
theta_1 = theta_1 -alpha * d_theta_1

print(f'theta_0_new : {theta_0}')
print('theta_1_new :', theta_1)
```

```python
theta_0 = 0
theta_1 = 2
alpha = 0.001

# Get number of data points
m = X.shape[0]

for i in range(1000):
    print(f'****************** Iteration {i} ********************\n')

    h = theta_0+theta_1*X
    print(f'h(x):{h}\n')

    e = h-y
    j=(e@e)/(2*m)
    print(f'Error Vector:\n{e}\n')
    print(f'j = {j}\n')

    d_theta_0 = e.sum()/m
    d_theta_1 = (e@X)/m
    grad = np.array([[d_theta_0],
                     [d_theta_1]])
    grad_norm = np.linalg.norm(grad)
    print(f'Gradient Vector:\n{grad}\n')
    print(f'Gradient Vector Norm:\n{grad_norm}\n')

    theta_0 = theta_0 - alpha * d_theta_0
    theta_1 = theta_1 - alpha * d_theta_1

    print(f'theta_0_new : {theta_0}')
    print(f'theta_1_new : {theta_1}\n')
```

```python
y_pred_GD_1 = theta_0 +theta_1 * X
print(f'h(x)= y_predict: \n {y_pred_GD_1}\n')
print(f'y_actual:\n{y}\n')
```

```python
plt.figure(figsize =(10,8))
plt.scatter(X,y)
plt.plot(X,y_pred_GD_1,'r*-')
plt.show()
```
