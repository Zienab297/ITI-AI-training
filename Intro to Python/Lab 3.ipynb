#Outliers detection

```python
sns.boxplot(x=df["housing_median_age"])
```

```python
sns.boxplot(x=df["longitude"])
```

```python
sns.boxplot(x=df["median_income"])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

#Visualization

```python
df.median_income.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Housing")
plt.ylabel('median_income')
plt.xlabel('Median income');
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['median_income'], df['households'])
ax.set_xlabel('Income')
ax.set_ylabel('household')
plt.show()
```
