##Checking for outliers and exculding them

```python
sns.boxplot(x=df['Price'])
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print(lower)
print(upper)
```

```python
lowdf = (df['Price'] < lower)
updf = (df['Price'] > upper)
```

```python
len(df['Price']) - (len(df['Price'][lowdf]) + len(df['Price'][updf]))
```

```python
df_filtered = df['Price'][~(lowdf | updf)]
df_filtered
```

```python
sns.boxplot(x=df_filtered)
```
